# Drop the NaN

```python
df_data_1 = df_data_1.dropna(how='any',axis=0)
```

```python
len(df_data_1)
```

```python
df_data_1['hasSleepApnea'].value_counts()
```

```python
df_data_1['hasSleepApnea'] = df_data_1['hasSleepApnea'].apply(lambda x:0 if x == False else 1)
```

```python
df_data_1['userId'].drop_duplicates()
```

```python
feature_cols = ['restingHeartRate','minutesAsleep', 'minutesAwake']
features_df = df_data_1[feature_cols]
```

```python
from sklearn import ensemble
from sklearn import model_selection
```

# From the Dataset , split the dataset into Testing and training

```python
fitbit_train, fitbit_test, target_train, target_test = model_selection.train_test_split(features_df, df_data_1.loc[:,'hasSleepApnea'], test_size=0.33, random_state=10)
```

# Train the Model and Get the accuracy

```python
clf1 = ensemble.GradientBoostingClassifier()
clf1.fit(fitbit_train, target_train)
clf1.score(fitbit_test,target_test)
```

# Use the model to Classify if the person is having Sleep Apnea or Not
